```python
import pandas as pd


station_file = pd.read_csv('station.csv')

weather_file = pd.read_csv('weather.csv')

trip_file = pd.read_csv('trip.csv',error_bad_lines = False)


```

## What is the average trip duration for a borrowed bicycle?

```python
trip_file['tripduration'].mean()
```

## What’s the most common age of a bicycle-sharer?

```python
# value_counts() function returns object containing counts of unique values. 
# The resulting object will be in descending order so that the first element is the most frequently-occurring element.
# -----------------------------------------------------------------------------------------------------------------
# idmax : function returns index of first occurrence of maximum over requested axis.
# While finding the index of the maximum value across any index, all NA/null values are excluded.

commonyear =trip_file['birthyear'].value_counts().idxmax()
age = 2020-commonyear
age
```

## Given all the weather data here, find the average precipitation per month, and the median precipitation.

```python
weather_file['Month']=pd.DatetimeIndex(weather_file['Date']).month
"""DatetimeIndex [source] Immutable ndarray of datetime64 data,  
and which can be boxed to Timestamp objects that are subclasses of datetime """
""" Exctract month and create a dedicated column df["Month"] from a 
#column in datetime format df["Date"]"""

""" Group by Month """
Month=weather_file.groupby('Month')['Precipitation_In']
#print(weather_file.groupby('Month'))
print(Month.mean())

print(Month.median())
```

## What’s the average number of bikes at a given bike station?

```python
""" isnull : returns a DataFrame object where all the values are replaced with 
a boolean value True for NULL values, and otherwise False """

date = station_file[station_file['decommission_date'].isnull()]
#print([station_file['decommission_date'].isnull()])
#print(station_file[station_file['decommission_date'].isnull()])
date['current_dockcount'].mean()
```

## When a bike station is modified, is it more likely that it’ll lose bikes or gain bikes? How do you know?

```python
""" dropna is used to remove rows and columns with Null/NaN values"""
"""sum :  It returns a series that contains the sum of all the values in each column"""
modified = station_file[['modification_date','install_dockcount','current_dockcount']].dropna()
print(modified['current_dockcount'].sum())
print(modified['install_dockcount'].sum())
modified['current_dockcount'].sum() - modified['install_dockcount'].sum()
```
